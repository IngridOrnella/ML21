```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
n_1 = 50
n_2 = 200
n_3 = 1000
```

```python
# normal
mu, sigma = 0, 1

def normal_samples(mu, sigma, n, N = 1000):
    x  = np.random.normal(mu, sigma, (N, n))
    return np.mean(x, axis = 1)
```

```python
res1 = normal_samples(mu, sigma, n_1)
res2 = normal_samples(mu, sigma, n_2)
res3 = normal_samples(mu, sigma, n_3)

plt.hist(res1)
plt.hist(res2)
plt.hist(res3)

plt.axvline(x=np.mean(res1))
plt.axvline(x=np.mean(res2))
plt.axvline(x=np.mean(res3))
```

```python
print(sigma ** 2)
print(np.var(res1))
print(np.var(res2))
print(np.var(res3))
```

```python
# bernoulli
p = 0.8

def binomial_samples(n, p, N = 1000):
    x  = np.random.binomial(n, p, N)/n
    return x
```

```python
res1 = binomial_samples(n_1, p)
res2 = binomial_samples(n_2, p)
res3 = binomial_samples(n_3, p)

plt.hist(res1)
plt.hist(res2)
plt.hist(res3)

plt.axvline(x=np.mean(res1))
plt.axvline(x=np.mean(res2))
plt.axvline(x=np.mean(res3))
```

```python
print(p * (1 - p))
print(np.var(res1))
print(np.var(res2))
print(np.var(res3))
```
